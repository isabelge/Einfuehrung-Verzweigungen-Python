Anstatt mehrere `if`s zu Verwenden gibt es auch die Möglichkeit `elif` zu nutzen. `elif` gehört zum gleichen Block wie das vorherige `if` und kann nicht alleine stehen. Es wird nur dann geprüft, wenn alle vorherigen Bedingungen falsch waren.

Das bedeutet, sobald ein `elif` wahr ist, werden alle weiteren `elif` und `else` übersprungen.

Wenn mehrere `if` nacheinander verwendet werden, so wird jede einzelne Bedingung geprüft.

- if -> startet einen neunen Verzweigungsblock
- elif -> weitere Abfrage, aber nur dann, wenn die vorherige Bedingung nicht erfüllt wurde
- else -> wird ausgeführt, wenn alle vorherigen Bedingungen nicht erfüllt wurden
